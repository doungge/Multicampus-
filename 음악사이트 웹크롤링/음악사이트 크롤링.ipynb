# 멜론 크롤링 결과를 엑셀로 저장하기

```python
# 멜론사이트 접속하기
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('C:/myexam/chromedrive/chromedriver.exe')
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 반복문을 이용해 곡가 가수명을 song_data에 저장하기
song_data = []
rank = 1

songs = soup.select('table > tbody > tr')
for song in songs:
    title = song.select('div.rank01 > span>a')[0].text
    singer = song.select('div.rank02>a')[0].text
    song_data.append(['Melon', rank, title, singer])
    rank = rank + 1
```

```python
# song_data 리스트를 이용해 데이터 프레임 만들기
import pandas as pd
columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, columns = columns)
pd_data.head()
```

```python
#크롤링 결과를 엑셀파일로 저장하기
pd_data.to_excel('melon.xlsx', index = False)
```

## 멜론 벅스 지니 크롤링 엑셀 파일 통합하기



```python
import pandas as pd
excel_names = ['melon.xlsx', 'bug.xlsx', 'genie.xlsx']

appended_data = pd.DataFrame()
for name in excel_names:
    pd_data = pd.read_excel(name)
    appended_data = appended_data.append(pd_data)
```

```python
#크롤링 결과 확인하기
appended_data.info()
```

```python
#통합한 크롤링 결과를 엑셀 파일로 저정하기
appended_data.to_excel('total.xlsx', index = False)
```
