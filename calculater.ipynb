## 3.2 거듭제곱과 나머지
 
  - 숫자A를 n번 곱하는 거듭제곱(Power)
  

```python
2*2*2*2*2
```

- 파이썬에서는 거듭제곱을 위한 연산자 **(두 별표 사이에는 공백이 없어야함)

```python
2**5
```

- 정수 뿐만 아니라 실수도 거듭제곱

```python
1.5**2
```

- 거듭제곱의 지수도 정수일 필요가 없음
- 파이썬에서 제공하는 수학(math) 라이브러리를 이용하면 거듭제공, 거듭제곱근 외에 다양한 수학 함수를 입력

```python
2**(1/2)
```

- 나머지를 구하기 위해 파이썬에서는 퍼센트 기호(%)를 이용
- 퍼센트 기호(%)를 나머지 연산자(Modulo operator)라고 함

```python
13%5
```

- 몫은 정수 나누기 연산자(//)로 구할 수 있음

```python
13//5
```

## 3.4 진수 표현과 변환

- 파이썬은 10진법 외에 2진법,8진법,16진법으로 숫자를 입출력하며 변환하는 방법을 제공
- 10 진수 이외에 2진수 8진수 16진 수를 인력하기 위해서는 숫자 앞에 각각 숫자앞에 `0b`,`0o`, `0x`를 붙임

```python
17
```

```python
0b10001
```

```python
0o21
```

```python
0x11
```

- 파이썬에서는 2진수, 8진수, 16진수 형태로 숫자를 입력해도 기본적으로 10진수로 출력
- 파이썬에서 10진수를 2진수, 8진수, 16진수로 변환하는 함수가 있고 각각의 함수는 `bin()`,`oct()`,`hex()`

```python
bin(17)
```

```python
oct(17)
```

```python
hex(17)
```

- 출력 결과는 숫자가 아니라 문자 열이라는 것임
- 출력 결과는 작음 따옴표('')안에 있음
- 출력 결과가 숫자가 아니라 문자열을 나타낸다.

- `bin()`, `oct()`, `hex()`함수를 이용한 결과는 산술 연산에 이용할 수 없음, 진수 변환은 연산이 모두 끝난 후에 해야함

```python
0b10+0o10+0x10-10
```

```python
bin(0b10 * 0x10 -10)
```

```python
oct(0b10 * 0x10 -10)
```

```python
hex(0b10 * 0x10 -10)
```


## 3.5 논리 연산 및 비교 연산

- 어떤 조건을 만족하는 참(true)과 만족하지 않는 거짓(False)를 이용하여 연산하는 논리 연산(logical operation), 논리 연산은 불린 연산(Boolean operation)이라고도 함


- 파이썬에서 논리 여산을 위한 데이터 타입은 불(bool)임 
- 불 데이터 타입에는 논리 참(true)혹은 논리 거짓(false)를 입력
-참(true) 혹은 거짓(false)를 입력할 때 참은 true 거짓은 False 를 입력


- 'true' 나 'false'처럼 따옴표를 사용하면 안됨.
- 따옴표를 이용해 입력하면 불 데이터가 아니라 문자열 데이터로 인식
- `True`를 true나 TRUE로 쓰거나 `False`를 false나 FALSE로 쓸 수 없음


```python
print(True)
```

```python
print(False)
```

 - 문자열을 출력하는 print('True')나 print('False')와 달리, 논리 연산의 결과를 출력한것
 - 불 데이터인 True나 False의 데이터 타입은 type() 함수를 이용해 확인


```python
type(True)
```

- 불 데이터의 경우 논리 연산만 할 수 있음
- 논리 연산에는 논리곱(and), 논리합(or) 논리 부정(not)등

- 논리곱(and)은 두 개의 불 데이터가 모두 참일 때만 참이고 나머지는 거짓
- 논리합(or)은 두 개의 불 데이터 중 하나라도 참이면 참이고 둘다 거짓이면 거짓
- 논리부정(not)은 하나의 불 데이터가 참이면 거짓이고 거짓이면 참

```python
print(True and False)
```

```python
print(True or False)
```

```python
print(not True)
```

 - 비교 연산자를 이용해 연산
 - 비교 연산의 결과는 불 데이터 형식으로 나옴
 

```python
print(5==3)
```

```python
print(5!=3)
```

```python
print(5<3)
```

```python
print(5>3)
```

```python
print(5<=3)
```

- 비교 연산자의 우선순위가 논리 연산자의 우선순위보다 높음
- 따라서 비교 연산자와 논리 연산자가 함꼐 있으면 비교 연산을 먼저함


```python
print(1>0 and -2<0)
```

- 비교 연산인 1>0과 -2<0을 먼저 수행하는데 결과는 모두 True
- 그 후 이 연산의 결과를 이용해 논리곱(and) 연산을 수행하는데 결과는 True

- 괄호와 연산이 함꼐 있으면 괄호의 우선순위가 높음
- 괄호가 여러 겹 있을 때는 가장 안쪽부터 먼저 계산함

```python
print((3<0) and ((-5>0)and (1>5)))
```

```python
print(3>0 and ((-5>0) and (1>5)))
```

```python
print(((3>0) or (-5>0))and ((4>8) or (3<0)))
```
